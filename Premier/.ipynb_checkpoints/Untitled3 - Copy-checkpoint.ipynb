```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {
    'Brighton and Hove Albion' : 'Brighton',
    'Huddersfield Town' : 'Huddersfield',
    'Manchester United' : 'Manchester Utd',
    'Newcastle United' : 'Newcastle Utd',
    'Sheffield United' : 'Sheffield Utd',
    'Tottenham Hotspur' : 'Tottenham',
    'West Bromwich Albion': 'West Brom',
    'West Ham United' : 'West Ham',
    'Wolverhampton Wanderers' : 'Wolves'}
```

```python
def rolling_averages( group, cols, new_cols) :
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(5, closed = 'left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
cols = ['poss','gf','ga','sh','sot','fk','pk','pkatt']
new_cols = [f'{c}_rolling' for c in cols]
```

```python
df_ml = df.groupby('team_name').apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
errors = []
misclassifications = []

for n in range(1,128):
    rfc = RandomForestClassifier(n_estimators=n, random_state=505, bootstrap= True, max_features= 'log2',min_samples_leaf = 2,min_samples_split= 30,oob_score= True)
    rfc.fit(train[param],train['result'])
    predss = rfc.predict(test[param])
    err = 1 - accuracy_score(predss,test['result'])
    n_missed = np.sum(preds != test['result']) # watch the video to understand this line!!
    errors.append(err)
    misclassifications.append(n_missed)
```
