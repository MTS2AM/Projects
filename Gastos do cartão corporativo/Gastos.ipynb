```python
df = pd.read_csv('Planilha.csv', header = 0, encoding='latin-1', sep=';')
```

```python
df= df.dropna()
```

```python
df['VALOR'] = df['VALOR'].apply(lambda x : str(x).replace('R$',' '))
```

```python
df.info()
```

```python
df['DATA PGTO'] = df['DATA PGTO'].astype('datetime64[ns]')
```

```python
df
```

```python
df['ANO'] = df['DATA PGTO'].dt.year
```

```python
def presidentes(ano) : 
    if ano >= 2003 and ano <= 2006 :
        return 'Lula 1'
    if ano >= 2007 and ano <= 2010 :
        return 'Lula 2'
    if ano >= 2011 and ano <= 2014 :
        return 'Dilma 1'
    if ano >= 2015 and ano <= 2016 :
        return 'Dilma 2'
    if ano >= 2017 and ano <= 2018 :
        return 'Temer'
    if ano >= 2019 and ano <= 2022 :
        return 'Bolsonaro'
```

```python
df['PRESIDENTES'] = df['ANO'].apply(presidentes)
```

```python
df['PRESIDENTES']
```

```python
df['VALOR'] = df['VALOR'].apply(lambda x : str(x).replace('.',''))
df['VALOR'] = df['VALOR'].apply(lambda x : str(x).replace(',','.'))
```
